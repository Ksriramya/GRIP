* Attacks per years

```python
x_year = terror['Year'].unique()
y_count_years = terror['Year'].value_counts().sort_index()
plt.figure(figsize = (15,10))
custom_colors = ["#7AB", "#EDA"]
# Bar plot with custom colors
sns.barplot(x=x_year, y=y_count_years, palette=custom_colors)
sns.barplot(x = x_year,
           y = y_count_years)
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack Vs Years')
plt.show()          
```

```python
# Total Number of people killed in terror attack
killData = terror.loc[:,'Killed']
print('Number of people killed by terror attack:', int(sum(killData.dropna())))# drop the NaN values
```

```python
# Let's look at what types of attacks these deaths were made of.
attackData = terror.loc[:,'AttackType']
# attackData
typeKillData = pd.concat([attackData, killData], axis=1)
```

```python
typeKillData.head()
```

```python
typeKillFormatData = typeKillData.pivot_table(columns='AttackType', values='Killed', aggfunc='sum')
typeKillFormatData
```

```python
typeKillFormatData.info()
```

```python
labels = typeKillFormatData.columns.tolist()
transpoze = typeKillFormatData.T
values = transpoze.values.flatten()  # Flatten the 2D array to a 1D array

fig, ax = plt.subplots(figsize=(8, 10), subplot_kw=dict(aspect="equal"))
plt.pie(values, labels=labels, startangle=90, autopct='%.2f%%')
plt.title('Types of terrorist attacks that cause deaths')
plt.show()
```

* Top Terrorist Groups Activities from 1970 to 2017

```python
groups_10 = terror[terror.Group.isin(terror.Group.value_counts()[1:11].index)]

# Use a custom color palette for a professional look
colors = sns.color_palette('Paired', 10)

pd.crosstab(groups_10.Year, groups_10.Group).plot(color=colors)
fig = plt.gcf()
fig.set_size_inches(18, 6)
plt.xticks(range(1970, 2017, 5))
plt.ylabel('Total Attacks')
plt.title('Top Terrorist Groups Activities from 1970 to 2017')
plt.legend(labels=['Al-Shabaab',
                   'Boko Haraam',
                   'FMLN',
                   'IRA',
                   'ISIL',
                   'PKK',
                   'NPA',
                   'FARC',
                   'SL',
                   'Taliban'], loc='upper left')
plt.show()
```

```python
group_attacks = terror.Group.value_counts().to_frame().drop('Unknown').reset_index()[:16]
group_attacks.columns = ['Terrorist Group', 'Total Attacks']
group_attacks
```

* Attacks by Terrorist Group

```python
group_attacks = terror.Group.value_counts().to_frame().drop('Unknown').reset_index()[:16]
group_attacks.columns = ['Terrorist Group', 'Total Attacks']
plt.subplots(figsize=(10,8))
sns.barplot(y=group_attacks['Terrorist Group'], x=group_attacks['Total Attacks'], palette='winter',
            edgecolor=sns.color_palette('dark',10))
# plt.xticks()
plt.title('Number Of Total Attacks by Terrorist Group')
plt.show()
```

```python
df_tal = terror[terror.Group == 'Taliban']
plt.figure(figsize=(13,7))
df_tal.groupby(['Year'])['Killed'].sum().plot(kind='bar',colormap='RdBu')
plt.title('People Killed by Taliban over the Years',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Number of people killed',fontsize=15)
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
